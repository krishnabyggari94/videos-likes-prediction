### Data Preprocessing

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('train.csv')
df_test=pd.read_csv('test.csv')
df.head(2)
```

```python
df_test.head(2)
```

```python
print(df.shape,df_test.shape)
```

```python
tot_df=pd.concat([df,df_test]).reset_index()
print(tot_df.shape)
tot_df.head(2)
```

```python
tot_df.drop(labels=['index','video_id'],axis=1,inplace=True)
tot_df.head(2)
```

```python
tot_df['likes'].fillna(value=-999,inplace=True)
tot_df['likes'].isnull().sum()
```

```python
tot_df['category_id'].value_counts()
```

```python
category_id_dict=dict(tot_df[tot_df['likes']!=-999].groupby('category_id')['likes'].median())
category_id_dict
```

```python
tot_df['category_id_median_likes']=tot_df['category_id'].replace(category_id_dict)
tot_df.head(2)
```

```python
chk=pd.get_dummies(tot_df['country_code'],drop_first=True)
tot_df=pd.concat([tot_df,chk],axis=1)
tot_df.head(2)
```

```python
tot_df['publish_date']=pd.to_datetime(tot_df['publish_date'],format='%Y-%m-%d')
tot_df['publish_year']=tot_df['publish_date'].dt.year
tot_df['publish_month']=tot_df['publish_date'].dt.month
tot_df['publish_day']=tot_df['publish_date'].dt.day
tot_df.head(2)
```

```python
from sklearn.preprocessing import StandardScaler
views_std=StandardScaler()
views_std.fit(tot_df[tot_df['likes']!=-999][['views']])
values=views_std.transform(tot_df[['views']])
tot_df=pd.concat([tot_df,pd.DataFrame(values,columns=['views_std'])],axis=1)
tot_df.head(2)
```

```python
dislikes_std=StandardScaler()
dislikes_std.fit(tot_df[tot_df['likes']!=-999][['dislikes']])
values=dislikes_std.transform(tot_df[['dislikes']])
tot_df=pd.concat([tot_df,pd.DataFrame(values,columns=['dislikes_std'])],axis=1)
tot_df.head(2)
```

```python
comments_std=StandardScaler()
comments_std.fit(tot_df[tot_df['likes']!=-999][['comment_count']])
values=comments_std.transform(tot_df[['comment_count']])
tot_df=pd.concat([tot_df,pd.DataFrame(values,columns=['comment_count_std'])],axis=1)
tot_df.head(2)
```

```python
category_id_median_likes_std=StandardScaler()
category_id_median_likes_std.fit(tot_df[tot_df['likes']!=-999][['category_id_median_likes']])
values=category_id_median_likes_std.transform(tot_df[['category_id_median_likes']])
tot_df=pd.concat([tot_df,pd.DataFrame(values,columns=['category_id_median_likes_std'])],axis=1)
tot_df.head(2)
```

```python
publish_year_std=StandardScaler()
publish_year_std.fit(tot_df[tot_df['likes']!=-999][['publish_year']])
values=publish_year_std.transform(tot_df[['publish_year']])
tot_df=pd.concat([tot_df,pd.DataFrame(values,columns=['publish_year_std'])],axis=1)

publish_month_std=StandardScaler()
publish_month_std.fit(tot_df[tot_df['likes']!=-999][['publish_month']])
values=publish_month_std.transform(tot_df[['publish_month']])
tot_df=pd.concat([tot_df,pd.DataFrame(values,columns=['publish_month_std'])],axis=1)

publish_day_std=StandardScaler()
publish_day_std.fit(tot_df[tot_df['likes']!=-999][['publish_day']])
values=publish_day_std.transform(tot_df[['publish_day']])
tot_df=pd.concat([tot_df,pd.DataFrame(values,columns=['publish_day_std'])],axis=1)

tot_df.head(2)
```

```python
tot_df.columns
```

```python
final_df=tot_df.drop(labels=['title','channel_title','category_id','publish_date','tags','views','dislikes','comment_count','description','country_code','category_id_median_likes','publish_year','publish_month','publish_day'],axis=1)
final_df.head(2)
```

```python
final_df.to_csv('final_df_without_text.csv',index=False)
tot_df.to_csv('tot_df.csv',index=False)
```

```python
test=final_df[final_df['likes']==-999]
test.head()
```

```python
train=final_df[final_df['likes']!=-999]
train.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train.drop(labels=['likes'],axis=1),train['likes'],test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_log_error
import math
rf_reg=RandomForestRegressor()
rf_reg.fit(x_train,y_train)
y_pred=rf_reg.predict(x_test)
error=1000.0*(math.sqrt(mean_squared_log_error(y_test,y_pred)))
print('Root Mean Squared Log Error is:',error)
```
